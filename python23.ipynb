```python
from abc import ABC, abstractmethod
class LibraryItem(ABC):
    def __init__(self, title, item_id):
        self.__title = title
        item_id.__item_id= item_id
        
    def get_title(self):
        return self.__title
        
    def set_title(self, title):
        self.__title = title

    def get_item_id(self):
        return self.__item_id

    def set_item_id(self, item_id):
        self.__item_id = item_id
        
    @abstractmethod
    def display_details(self):
        pass
```

```python
class Book(LibraryItem):
    def __init__(self, title, item_id, author, isbn, publication_year):
        super().__init__(title, item_id)
        self.author = author
        self.isbn = isbn
        self.publication_year = publication_year

    def display_details(self):
        print("Book Details:")
        print(f"Title: {self.get_title()}")
        print(f"Item ID: {self.get_item_id()}")
        print(f"Author: {self.author}")
        print(f"ISBN: {self.isbn}")
        print(f"Publication Year: {self.publication_year}")
```

```python
class Magazine(LibraryItem):
    def __init__(self, title, item_id, issue_number, publisher):
        super().__init__(title, item_id)
        self.issue_number = issue_number
        self.publisher = publisher

    def display_details(self):
        print("Magazine Details:")
        print(f"Title: {self.get_title()}")
        print(f"Item ID: {self.get_item_id()}")
        print(f"Issue Number: {self.issue_number}")
        print(f"Publisher: {self.publisher}")
        
```

```python
class LibraryMember:
    def __init__(self, member_id, name):
        self.member_id = member_id
        self.name = name
        self.borrowed_items = []

    def borrow_item(self, item):
        self.borrowed_items.append(item)
        print(f"{self.name} borrowed: {item.get_title()}")

    def display_borrowed_items(self):
        print(f"{self.name}'s Borrowed Items:")
        for item in self.borrowed_items:
            item.display_details()
```

```python

book = Book("tech245", "B1220567", "Simron", "1234567890", 2004)
magazine = Magazine("Sukrabar", "M1", "Issue 24", "Publishers")


member = LibraryMember("M001", "Alice")
member.borrow_item(book)
member.borrow_item(magazine)

# Show borrowed items
member.display_borrowed_items()
```
